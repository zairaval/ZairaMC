2-. Utilizando los nombres del archivo .mat, muestra el grafo en 4 formas distintas (shell debe ser una de ellas)

```python
import networkx as nx

# Crear el grafo a partir de la matriz de adyacencia y los nombres de los nodos
G = nx.from_numpy_array(CIJ)
node_labels = {i: name[0] for i, name in enumerate(mat_data['Names'].flatten())}
G = nx.relabel_nodes(G, node_labels)


# Configuración del gráfico
fig, axs = plt.subplots(2, 2, figsize=(15, 15))

# Disposiciones distintas para visualizar el grafo
layouts = [
    ('spring', nx.spring_layout(G)),
    ('shell', nx.shell_layout(G)),
    ('circular', nx.circular_layout(G)),
    ('spectral', nx.spectral_layout(G))
]

# Graficar el grafo en distintas disposiciones
for ax, (layout_name, layout) in zip(axs.flatten(), layouts):
    nx.draw(G, pos=layout, with_labels=True, node_size=500, font_size=10, ax=ax)
    ax.set_title(f"Grafo en disposición {layout_name}")

plt.tight_layout()
plt.show()

```

3-. Determina el nodo con mayores conexiones


```python
# Obtener el grado de cada nodo
node_degrees = dict(G.degree())

# Encontrar el nodo con el mayor número de conexiones
max_degree_node = max(node_degrees, key=node_degrees.get)
max_degree = node_degrees[max_degree_node]

print(f"El nodo con mayores conexiones es '{max_degree_node}' con un grado de {max_degree}.")
```

4-. Muestra el grado solo destacando las conexiones del nodo encontrado en el ejercicio anterior

```python
# Crear un color diferente para las conexiones del nodo con mayor grado
node_colors = ['red' if node == max_degree_node else 'lightblue' for node in G.nodes()]
edge_colors = ['red' if max_degree_node in edge else 'gray' for edge in G.edges()]

# Dibujar el grafo
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=500, node_color=node_colors, edge_color=edge_colors, font_size=10)
plt.title(f"Grafo destacando conexiones del nodo '{max_degree_node}'")
plt.show()
```

3-. Genera un array de 1000 números con distribución normal, con mu= ej1, y sigma= ej2

```python
import numpy as np

# Parámetros de la distribución
mu = 0  # Media de la distribución
sigma = 1  # Desviación estándar de la distribución

# Generar el array de 1000 números
array_normal = np.random.normal(mu, sigma, 1000)
array_normal
```

4-. Plotear la distribución de los 1000 números


```python
# Crear el histograma de los datos
plt.hist(array_normal, bins=30, density=True, alpha=0.6, color='blue')

# Graficar la curva de la distribución normal
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = (1 / (np.sqrt(2 * np.pi * sigma**2))) * np.exp(- (x - mu)**2 / (2 * sigma**2))
plt.plot(x, p, 'k', linewidth=2)

plt.xlabel("Valor")
plt.ylabel("Densidad de probabilidad")
plt.title("Distribución normal de 1000 números")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def generar_array(mu, sigma, total_de_numeros):
    return np.random.normal(mu, sigma, total_de_numeros)

def calcular_sigmas(array):
    mu = np.mean(array)
    sigma = np.std(array)
    return mu, sigma

def plotear_distribucion(array, mu, sigma):
    plt.hist(array, bins=30, alpha=0.7, color='blue', edgecolor='black')
    plt.title(f'Distribución Normal con $\\mu={mu}$ y $\\sigma={sigma}$')
    plt.xlabel('Valor')
    plt.ylabel('Frecuencia')
    plt.show()

# Ejemplo de uso:
mu = 0
sigma = 1
total_de_numeros = 1000
array_numeros = generar_array(mu, sigma, total_de_numeros)
mu_calculado, sigma_calculado = calcular_sigmas(array_numeros)
plotear_distribucion(array_numeros, mu_calculado, sigma_calculado)
print(f'\\mu calculado: {mu_calculado}, \\sigma calculado: {sigma_calculado}')
```

```python
def calcular_porcentajes(array, mu, sigma): 
    rango1 = np.sum(np.abs(array - mu) <= sigma) / len(array) * 100 
    rango2 = np.sum(np.abs(array - mu) <= 2 * sigma) / len(array) * 100 
    rango3 = np.sum(np.abs(array - mu) <= 3 * sigma) / len(array) * 100 
    return rango1, rango2, rango3
```

9-. Usar la función anterior con los valores (0, 1., 1000), (3, 5, 5000), (-3, 3, 10000)

Valores (0, 1., 1000)

```python
# Conjunto 1
mu1, sigma1, total1 = 0, 1, 1000

array1 = generar_array(mu1, sigma1, total1)

mu1_calculado, sigma1_calculado = calcular_sigmas(array1)

plotear_distribucion(array1, mu1_calculado, sigma1_calculado, f'Distribución Normal con $\\mu={mu1}$ y $\\sigma={sigma1}$')

rango1_1, rango2_1, rango3_1 = calcular_porcentajes(array1, mu1_calculado, sigma1_calculado)
```

Valores (3, 5, 5000)

```python
#Conjunto 2 
mu2, sigma2, total2 = 3, 5, 5000 

array2 = generar_array(mu2, sigma2, total2) 

mu2_calculado, sigma2_calculado = calcular_sigmas(array2) 

plotear_distribucion(array2, mu2_calculado, sigma2_calculado, f'Distribución Normal con $\\mu={mu2}$ y $\\sigma={sigma2}$') 

rango1_2, rango2_2, rango3_2 = calcular_porcentajes(array2, mu2_calculado, sigma2_calculado)
```

Valores (-3, 3, 10000)

```python
# Conjunto 3 
mu3, sigma3, total3 = -3, 3, 10000 

array3 = generar_array(mu3, sigma3, total3) 

mu3_calculado, sigma3_calculado = calcular_sigmas(array3) 

plotear_distribucion(array3, mu3_calculado, sigma3_calculado, f'Distribución Normal con $\\mu={mu3}$ y $\\sigma={sigma3}$') 

rango1_3, rango2_3, rango3_3 = calcular_porcentajes(array3, mu3_calculado, sigma3_calculado)
```
